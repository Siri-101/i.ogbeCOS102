```python
import math
import numpy as np

answer = input("Quadratic (A), Quartic (B), Cubic (C)")

if answer == "A":
    print("Ax^2 + Bx^1 + Cx")
    variable = input("A, B, C: ")
    array = []
    
    for x in variable.split(","):
        array.append(int(x))
    discriminant = (array[1] ** 2) - (4 * array[0] * array[2])
    if discriminant < 0:
        print("Error: discriminant less than 0")
    else: 
        root1 = (-array[1] + math.sqrt(discriminant)) / (2 * array[0])
        root2 = (-array[1] - math.sqrt(discriminant)) / (2 * array[0])
        print(f"root1 and root2: {root1}, {root2}")

if answer == "B":
    print("Ax^3 + Bx^2 + Cx + D")
    variable = input("A, B, C, D: ")
    array = []
    
    for x in variable.split(","):
        array.append(int(x))
    
    roots = np.roots(array)
    print(roots)

if answer == "C":
    print("Ax^4 + Bx^3 + Cx^2 + Dx + E")
    variable = input("A, B, C, D, E: ")
    array = []
    
    for x in variable.split(","):
        array.append(int(x))
    roots = np.roots(array)
    print(roots)
```
