```python
user_input = input("What would you like to calculate? Total amount paid back in \n S: Simple Interest \n C: Compound Interest \n A: Annuity Plan \n Type S, C or A: ")

def validation(num):
    try:
        num = float(num)
        return num
    except ValueError:
        print("Nan")

def compound_interest():
    principal = input("What is the Principal: ")
    principal = validation(principal)
    rate = input("What is the interest rate of the loan _____%: ")
    rate = validation(rate)
    rate_to_fraction = rate/100
    duration = input("Duration of the loan in years: ")
    duration = validation(duration)
    num = input("How many times yearly will it be compounded? ")
    num = validation(num)
    c_amount = principal * ((1 + (rate_to_fraction/num))**(num*duration))
    print(f"Amount to be payed back after {duration}: {c_amount}")


def simple_interest():
    principal = input("What is the Principal: ")
    principal = validation(principal)
    rate = input("What is the interest rate of the loan _____%: ")
    rate = validation(rate)
    rate_to_fraction = rate/100
    duration = input("Duration of the loan in years: ")
    duration = validation(duration)
    s_amount = principal * (1 + (rate_to_fraction * duration))
    print(f"Amount to be payed back after {duration}: {s_amount}")

def annuity_plan():
    pmt = input("What is the Amount to be payed periodically: ")
    pmt = validation(pmt)
    rate = input("What is the interest rate of the loan _____%: ")
    rate = validation(rate)
    rate_to_fraction = rate/100
    duration = input("Duration of the loan in years: ")
    duration = validation(duration)
    num = input("How many times yearly will it be compounded? ")
    num = validation(num)
    a_amount = pmt * ((1 + (rate_to_fraction/num))**(num*duration)-1)/(rate_to_fraction/num)
    print(f"Amount to be payed back after {duration}: {a_amount}")



if user_input.lower() == "c" :
    compound_interest()
elif user_input.lower() == "s" :
    simple_interest()
elif user_input.lower() == "a" :
    annuity_plan()



    
    
    
```
