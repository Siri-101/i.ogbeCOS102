```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')


titanic = pd.read_csv('C:/Users/New/Downloads/titanic.csv')

cleaned1 = titanic.drop(columns = ['Name', 'Ticket','Age', 'Cabin','Embarked', 'PassengerId'])
cleaned1['Sex'] = cleaned1['Sex'].replace({'male': 0, 'female': 1})
cleaned1['Pclass'] = cleaned1['Pclass'].replace({1: 3, 3: 1})
cleaned1 = cleaned1.dropna()
cleaned1.isnull().sum()

corr = cleaned1.corr(method = 'kendall')
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True), 
cleaned1.columns

```

From the heat map above, It is evident that there is correlation between the passenger class and survival. I.e. People who were in First class had higher chances of surviving and this reflected with the relationship between fare payed and survival. First-class who paid higher had higher chance of survival and thus the higher the fare payed, they higher the survival rate.
It is also evident that the sex of the passenger had no particular determinant of the class of residence where they stayed within the ship
However, Sex of the passenger had an influence on wether or not they survived. From the map, it is evident that women had a higher chance of surviving
