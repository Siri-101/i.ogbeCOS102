```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

file = pd.read_csv('C:/Users/New/Downloads/AB_NYC_2019.csv')
file.isnull().sum()
```

```python
file = file.drop(columns = ['name'])
file.isnull().sum()
```

```python

file.dropna(how='any',inplace=True)
file.isnull().sum()
```

```python
file.columns
```

```python
num_value = file.drop(columns = ['neighbourhood_group', 'neighbourhood',  'name', 'host_name', 'last_review'])
num_value['room_type'] = num_value['room_type'].replace({'Private room': 3, 'Entire home/apt': 6, 'Shared room': 1})
    
corr = num_value.corr(method='kendall')
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True), 
num_value.columns
```

Evidentlt, price was largely influenced by room-type which was graded from 1 to 3, based on if it is the entire home, private room or shared room
Number of reviews and reviews per month have high correlation

Reviews per month and availability throughout the year also have correlation, likely because with higher availability and hence higher boarders who could leave higher reviews
